```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/drive/MyDrive/Data/House Price India.csv")
pd.plotting.scatter_matrix(data.loc[:, "id":"Price"], diagonal="Assignment",figsize=(20,15))
plt.show()
```

```python
sns.lmplot(x ='number of bedrooms', y ='number of bathrooms',fit_reg = False, hue = 'Date',data = data)
```

```python
data_file = pd.DataFrame(np.random.normal(size=100000))

data_file.plot(kind="density",
              figsize=(10,10));


plt.vlines(data_file.mean(),     # Plot black line at mean
           ymin=0, 
           ymax=0.4,
           linewidth=5.0);

plt.vlines(data_file.median(),   # Plot red line at median  
           ymin=0, 
           ymax=0.4, 
           linewidth=2.0,
           color="red");
plt.vlines(data_file.std(),
           ymin=0, 
           ymax=0.4, 
           linewidth=2.0,
           color="teal");
```
